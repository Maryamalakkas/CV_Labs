```python
# Import necessary libraries
import cv2
import matplotlib.pyplot as plt

# Load the image from the specified path
image_path = r"C:\Users\Admin\Computer-Vision\Lab-5-done\Statue-of-Liberty-Island-New-York.jpeg"
image = cv2.imread(image_path)

# Resize the image to a smaller size
width = 512  # desired width
height = 512  # desired height
resized_image = cv2.resize(image, (width, height))

# Convert the resized image to grayscale
gray = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

# Print the grayscale image array
print(gray)

# Display the grayscale image using matplotlib
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')  # Hide the axis
plt.show()


```

## Feature Extraction Using SIFT
In this code snippet, we demonstrate how to use the SIFT (Scale-Invariant Feature Transform) algorithm for feature extraction in an image. We begin by importing the necessary libraries, OpenCV for computer vision tasks and matplotlib for visualizing the results. Next, we create a SIFT detector object using cv2.SIFT_create(). The detectAndCompute function is employed to detect keypoints and compute the corresponding descriptors in the grayscale image. SIFT is a powerful algorithm for detecting and describing local features in images, making it invariant to scale, rotation, and illumination changes. It is widely used for various computer vision applications such as object recognition, image stitching, and 3D reconstruction. After detecting the keypoints, we visualize them on the grayscale image using the cv2.drawKeypoints function. Finally, we display the image with the detected keypoints using matplotlib to provide a clear visual representation of where SIFT has identified significant features in the image.

```python
# Import necessary libraries
import cv2
import matplotlib.pyplot as plt

# Create a SIFT (Scale-Invariant Feature Transform) detector object
sift = cv2.SIFT_create()

# Detect keypoints and compute descriptors using SIFT
# The detectAndCompute function finds keypoints and descriptors in the image
keypoints_sift, descriptors_sift = sift.detectAndCompute(gray, None)

# Print the number of keypoints detected in the image
print("Number of Keypoints (SIFT):", len(keypoints_sift))

# Draw the detected keypoints on the grayscale image
# The drawKeypoints function visualizes the keypoints on the image
image_with_keypoints_sift = cv2.drawKeypoints(gray, keypoints_sift, None)

# Display the image with the keypoints using matplotlib
plt.imshow(image_with_keypoints_sift, cmap='gray')
plt.title('Image with SIFT Keypoints')
plt.axis('off')  # Hide the axis
plt.show()

```

## Feature Extraction Using ORB
In this code snippet, we demonstrate how to use the ORB (Oriented FAST and Rotated BRIEF) algorithm for feature extraction in an image. We start by importing the necessary libraries, including OpenCV for computer vision tasks and matplotlib for visualizing the results. We then create an ORB detector object using cv2.ORB_create(). The detectAndCompute function is used to detect keypoints and compute the corresponding descriptors in the grayscale image. ORB is a robust feature detection and description algorithm that combines the FAST keypoint detector and the BRIEF descriptor, making it efficient for real-time applications. After detecting the keypoints, we visualize them on the grayscale image using the cv2.drawKeypoints function. Finally, we display the image with the detected keypoints using matplotlib to provide a clear visual representation of where ORB has identified significant features in the image. This technique is particularly useful for tasks such as object recognition, image matching, and tracking.

```python
# Import necessary libraries
import cv2
import matplotlib.pyplot as plt

# Create an ORB (Oriented FAST and Rotated BRIEF) detector object
orb = cv2.ORB_create()

# Detect keypoints and compute descriptors using ORB
# The detectAndCompute function finds keypoints and descriptors in the image
keypoints_orb, descriptors_orb = orb.detectAndCompute(gray, None)

# Print the number of keypoints detected in the image
print("Number of Keypoints (ORB):", len(keypoints_orb))

# Draw the detected keypoints on the grayscale image
# The drawKeypoints function visualizes the keypoints on the image
image_with_keypoints_orb = cv2.drawKeypoints(gray, keypoints_orb, None)

# Display the image with the keypoints using matplotlib
plt.imshow(image_with_keypoints_orb, cmap='gray')
plt.title('Image with ORB Keypoints')
plt.axis('off')  # Hide the axis
plt.show()

```

## Feature Extraction Using BRISK
In this code snippet, we demonstrate how to use the BRISK (Binary Robust Invariant Scalable Keypoints) algorithm for feature extraction in an image. Initially, we import the necessary libraries, including OpenCV for computer vision tasks and matplotlib for visualizing the results. We then initialize the BRISK detector using cv2.BRISK_create(). The detectAndCompute function is employed to detect keypoints and compute the corresponding descriptors in the grayscale image. Keypoints are distinctive points in the image, such as corners or blobs, which BRISK identifies reliably across different scales and rotations. After detecting the keypoints, we use the cv2.drawKeypoints function to visualize these keypoints on the grayscale image. Finally, we display the image with the detected keypoints using matplotlib to provide a clear visual representation of where BRISK has identified significant features in the image. This approach is valuable for tasks such as object recognition, image stitching, and 3D reconstruction.

```python
# Import necessary libraries
import cv2
import matplotlib.pyplot as plt

# Initialize the BRISK (Binary Robust Invariant Scalable Keypoints) detector
brisk = cv2.BRISK_create()

# Detect keypoints and compute descriptors using BRISK
# The detectAndCompute function finds keypoints and descriptors in the image
keypoints, descriptors = brisk.detectAndCompute(gray, None)

# Draw the detected keypoints on the grayscale image
# The drawKeypoints function visualizes the keypoints on the image
image_with_keypoints = cv2.drawKeypoints(gray, keypoints, None)

# Display the image with the keypoints using matplotlib
plt.imshow(image_with_keypoints, cmap='gray')
plt.title('Image with BRISK Keypoints')
plt.axis('off')  # Hide the axis
plt.show()
```
